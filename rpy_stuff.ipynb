```python
import rpy2
```

```python
import pandas as pd
```

```python
sleep_df = pd.read_csv('data/sleepstudy.csv')
```

```python
from rpy2 import robjects as ro
from rpy2.robjects import pandas2ri
from rpy2.robjects.packages import importr

pandas2ri.activate()
R = ro.r
lme = importr('lme4')
base = importr('base')
```

```python
def multilevel_linear_regression(formula, data):

    return lme.lmer(formula, data=data)

def multilevel_logistic_regression(formula, data):

    return lme.glmer(formula, family='binomial', data=data)
```

```python
M = multilevel_linear_regression('Reaction ~ Days + (1|Subject)', sleep_df)
#M_null = multilevel_logistic_regression('y ~ 1 +' + random_effects, data)
```

```python
print(base.summary(M))
```
