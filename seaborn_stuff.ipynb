```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
weight_df = pd.read_csv('weight.csv')
```

```python
sns.scatterplot(x = 'height', y = 'weight', data = weight_df)
```

```python
sns.scatterplot(x = 'height', y = 'weight', hue = 'gender', data = weight_df, s = 100)
```

```python
sns.relplot(x="height", y="weight", hue="gender", col = 'race', col_wrap=2, data=weight_df)
```

```python
sns.lmplot(x = 'height', y = 'weight', hue = 'gender', row = 'handedness', 
           #col_wrap=1,
           scatter_kws={'s': 10},
           ci = None,
           data = weight_df.query('(handedness == "left") | (handedness == "right")'))
```

```python
sns.catplot(x="gender", y="height", kind = 'box', col = 'handedness', data = weight_df)
```
